```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python


# Set the path to the file you'd like to load
file_path = "/kaggle/input/horror-movies-profits-dataset/movie_profit.csv"

# Load the latest version
df = pd.read_csv(file_path)

df.head()
```

```python
df.sort_values("production_budget",ascending=False)
```

```python
df.groupby("mpaa_rating")["domestic_gross"].mean().sort_values(ascending=True)
```

```python
plt.hist(df["worldwide_gross"])
plt.title("Distribution of worldwide_gross")
plt.xlabel("Frequency")
plt.ylabel("Worldwide Gross")
plt.show()
```

```python
plt.bar(df["release_date"].head(),df["domestic_gross"].head())
plt.title("Release date Domestic Gross")
plt.xlabel("release Date")
plt.ylabel("Domestic Gross")
plt.show()
```

```python
df.groupby("distributor")["worldwide_gross"].sum().sort_values(ascending=False)
```

```python
df.groupby("genre")["production_budget"].mean()
```

```python
df.groupby(df["release_date"].str.slice(6,8))["movie"].count().head()
```

```python
df.groupby("genre")["domestic_gross"].max()
```

```python
df.groupby("distributor")["worldwide_gross"].mean()
```

```python
df.groupby(df["release_date"].str.slice(6,8))["worldwide_gross"].mean().plot(kind="line")
plt.title("Mean World Gross Over Years")
plt.xlabel("Release Year")
plt.ylabel("Mean Worldwide Gross")
plt.show()
```

```python
df.groupby("mpaa_rating")["domestic_gross"].sum().plot(kind="pie")
plt.title("Total Domestic Gross by Rating")
plt.xlabel("")
plt.ylabel("")
plt.show()
```
